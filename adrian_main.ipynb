```python
# Crear una lista con todas las localizaciones (las que estaban en tu diccionario)
localizaciones = [
    "Total Nacional", "Andalucía", "Aragón", "Asturias, Principado de", "Balears, Illes", "Canarias", 
    "Cantabria", "Castilla y León", "Castilla - La Mancha", "Cataluña", "Comunitat Valenciana", 
    "Extremadura", "Galicia", "Madrid, Comunidad de", "Murcia, Región de", "País Vasco", "Rioja, La"
]

# Crear una lista con los rangos de edad
edades = ["Todas las edades", "Menos de 25 años", "De 25 a 34 años", "De 35 a 44 años", "De 45 a 54 años", "55 y más años"]

# Crear dos nuevas columnas: una para 'Localización' y otra para 'Rango de edad'
# Procesamos primero las localizaciones
df_salario_ccaa_sexo_y_edad['Localización'] = df_salario_ccaa_sexo_y_edad.iloc[:, 0].apply(lambda x: x if x in localizaciones else None)

# Rellenar las celdas de Localización hacia adelante (hacia abajo)
df_salario_ccaa_sexo_y_edad['Localización'].fillna(method='ffill', inplace=True)

# Luego procesamos los rangos de edad, ignorando las localizaciones ya procesadas
df_salario_ccaa_sexo_y_edad['Rango de edad'] = df_salario_ccaa_sexo_y_edad.iloc[:, 0].apply(lambda x: x if x in edades else None)

# Ahora podemos eliminar la primera columna y crear el MultiIndex
df_salario_ccaa_sexo_y_edad.drop(df_salario_ccaa_sexo_y_edad.columns[0], axis=1, inplace=True)

# Aplicar el MultiIndex con las nuevas columnas 'Localización' y 'Rango de edad'
df_salario_ccaa_sexo_y_edad = df_salario_ccaa_sexo_y_edad.set_index(['Localización', 'Rango de edad'])

# El dataframe ahora tendrá un MultiIndex adecuado
```

```python
df_salario_ccaa_sexo_y_edad.head(10)
```

```python
# Lista de reemplazo
edades = [np.nan, "Todas las edades", "Menos de 25 años", "De 25 a 34 años", "De 35 a 44 años", "De 45 a 54 años", "55 y más años"]

# Crear una columna temporal con el índice cíclico
df_salario_ccaa_sexo_y_edad.reset_index(inplace=True)  # Convertir MultiIndex a columnas normales
cycle_length = len(edades)
replacement_indices = np.arange(len(df_salario_ccaa_sexo_y_edad)) % cycle_length
df_salario_ccaa_sexo_y_edad['Replacement Index'] = replacement_indices

# Reemplazar los valores en la columna 'Rango de edad'
df_salario_ccaa_sexo_y_edad['Rango de edad'] = df_salario_ccaa_sexo_y_edad['Replacement Index'].map(lambda x: edades[x])

# Volver a establecer el MultiIndex
df_salario_ccaa_sexo_y_edad.set_index(['Localización', 'Rango de edad'], inplace=True)  # Cambia 'level_0' por el primer nivel original del índice

# Eliminar la columna temporal
df_salario_ccaa_sexo_y_edad.drop(columns=['Replacement Index'], inplace=True)

df_salario_ccaa_sexo_y_edad
```
