# Tablas del INE: IPC y salarios

```python
df_salario_nacional_sexo_y_edad_clean = pd.read_excel("df_salario_nacional_sexo_y_edad_clean.xlsx")
df_salario_nacional_sexo_y_edad_clean
```

```python
df_salario_INE = df_salario_nacional_sexo_y_edad_clean.head(1)
```

```python
df_IPC_INE = ipc_ccaa_media_anual_2002_2023.head(1)
df_IPC_INE
```

```python
df_IPC_INE_2008 = df_IPC_INE.loc[:, '2008':'2023']
df_IPC_INE_2008
```

```python
df_combinado = pd.concat([df_salario_INE, df_IPC_INE_2008], axis=0)
df_combinado.rename({0: 'Salarios', 1: 'IPC'}, inplace=True)
df_combinado = df_combinado.drop(df_combinado.columns[0], axis=1)
df_combinado
```

```python
df_combinado.loc["Salarios", "2022"] = 26948.87
df_combinado_t = df_combinado.T
df_combinado_t
```

```python
df_combinado_t.dtypes
```

```python
df_combinado_t.index = df_combinado_t.index.astype(int)
```

```python
df_combinado_t['Variación Anual'] = df_combinado_t['Salarios'].pct_change() * 100
```

```python
media_variacion_anual = df_combinado_t['Variación Anual'].mean()
```

```python
salario_2022 = df_combinado_t.loc[2022, 'Salarios']
```

```python
salario_estimado_2023 = round(salario_2022 * (1 + media_variacion_anual / 100), 2)
```

```python
df_combinado_t.loc[2023, 'Salarios'] = salario_estimado_2023
```

```python
df_combinado_t
```

```python
# Calcular el poder adquisitivo (Salarios ajustados por IPC)
df_combinado_t['Poder Adquisitivo'] = df_combinado_t['Salarios'] / df_combinado_t['IPC']
df_combinado_t
```

```python
import matplotlib.pyplot as plt

# Crear la figura y los ejes
plt.figure(figsize=(10, 6))

# Graficar el poder adquisitivo
plt.plot(df_combinado_t.index, df_combinado_t['Poder Adquisitivo'], marker='o', label='Poder Adquisitivo')

# Resaltar los últimos 5 años
plt.plot(df_combinado_t.index[-5:], df_combinado_t['Poder Adquisitivo'][-5:], color='red', marker='o', linestyle='--', label='Últimos 5 años')

# Etiquetas y título
plt.title('Evolución del Poder Adquisitivo (Ajustado por IPC)', fontsize=14)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Poder Adquisitivo (Salario / IPC)', fontsize=12)
plt.grid(True)

# Añadir leyenda
plt.legend()

# Mostrar el gráfico
plt.show()
```

```python
import numpy as np

# Calcular la tasa de crecimiento anual de los salarios e IPC
df_combinado_t['Crecimiento Salarial'] = df_combinado_t['Salarios'].pct_change() * 100
df_combinado_t['Crecimiento IPC'] = df_combinado_t['IPC'].pct_change() * 100

# Comprobar si en algún año el crecimiento del IPC es al menos el doble que el de los salarios
df_combinado_t['Doble IPC'] = np.where(df_combinado_t['Crecimiento IPC'] >= 2 * df_combinado_t['Crecimiento Salarial'], True, False)

# Verificar si existe algún año donde el IPC dobló al crecimiento salarial
resultado = df_combinado_t[df_combinado_t['Doble IPC'] == True]

if not resultado.empty:
    print("En los siguientes años, el crecimiento del IPC fue al menos el doble que el crecimiento salarial:")
    print(resultado[['Crecimiento Salarial', 'Crecimiento IPC']])
else:
    print("No hay ningún año en que el crecimiento del IPC haya doblado al crecimiento salarial.")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Calcular la tasa de crecimiento anual de los salarios e IPC
df_combinado_t['Crecimiento Salarial'] = df_combinado_t['Salarios'].pct_change() * 100
df_combinado_t['Crecimiento IPC'] = df_combinado_t['IPC'].pct_change() * 100

# Comprobar si en algún año el crecimiento del IPC es al menos el doble que el de los salarios
df_combinado_t['Doble IPC'] = np.where(df_combinado_t['Crecimiento IPC'] >= 2 * df_combinado_t['Crecimiento Salarial'], True, False)

# Crear el gráfico
plt.figure(figsize=(10, 6))

# Graficar la tasa de crecimiento salarial
plt.plot(df_combinado_t.index, df_combinado_t['Crecimiento Salarial'], marker='o', label='Crecimiento Salarial (%)')

# Graficar la tasa de crecimiento del IPC
plt.plot(df_combinado_t.index, df_combinado_t['Crecimiento IPC'], marker='o', color='red', label='Crecimiento IPC (%)')

# Resaltar los años donde el IPC fue al menos el doble del crecimiento salarial
doble_ipc_years = df_combinado_t[df_combinado_t['Doble IPC'] == True].index
plt.scatter(doble_ipc_years, df_combinado_t.loc[doble_ipc_years, 'Crecimiento IPC'], color='orange', label='IPC > 2x Crec. Salarial', zorder=5)

# Etiquetas y título
plt.title('Comparación Crecimiento Salarial vs IPC (2008-2023)', fontsize=14)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Tasa de Crecimiento (%)', fontsize=12)
plt.grid(True)

# Añadir leyenda
plt.legend()

# Mostrar el gráfico
plt.show()
```

# Comparativa desviaciones típicas salarios INE vs InfoJobs

### Informe InfoJobs 2022

```python
# Informe InfoJobs 2022
salarios_ij_2022
```

```python
salarios_ij_2022_hasta2021 = salarios_ij_2022.iloc[:-1].copy()
salarios_ij_2022_hasta2021
```

```python
salarios_ij_2022_hasta2021.loc[:, 'Tasa de Crecimiento Anual (%)'] = (salarios_ij_2022_hasta2021['Salario Promedio Ofertado (€)'].pct_change() * 100)

tasas_de_crecimiento_IJ = salarios_ij_2022_hasta2021['Tasa de Crecimiento Anual (%)'].dropna()

tasas_de_crecimiento_IJ
```

### Datos INE

```python
salario_INE_nacional_edad = pd.read_excel("databases/salarios/salario_nacional_sexo_y_edad/28189.xlsx")
salario_INE_nacional_edad
```

```python
años = salario_INE_nacional_edad.loc[6, 'Unnamed: 1':'Unnamed: 14'].values
años
```

```python
fila_todas_edades = salario_INE_nacional_edad.loc[7, 'Unnamed: 1':'Unnamed: 14']
fila_todas_edades
```

```python
fila_todas_edades.index = años
fila_todas_edades
```

```python
salarios_2011_2021 = fila_todas_edades.loc[2011:2021]
salarios_2011_2021
```

```python
tasas_crecimiento_anual = salarios_2011_2021.pct_change() * 100
tasas_crecimiento_anual = tasas_crecimiento_anual.dropna()

tasas_crecimiento_anual
```

### Medidas de estadística descriptiva

```python
#InfoJobs

media_tasas_crecimiento_IJ = tasas_de_crecimiento_IJ.mean()
desviacion_estandar_IJ = tasas_de_crecimiento_IJ.std()

print("La media de las tasas de crecimiento anual es:", media_tasas_crecimiento_IJ)
print("Desviación Estándar de las Tasas de Crecimiento Anual:", desviacion_estandar_IJ)
```

```python
#INE

media_tasas_crecimiento_INE = tasas_crecimiento_anual.mean()
desviacion_tasas_crecimiento_INE = tasas_crecimiento_anual.std()

print(f"La media de las tasas de crecimiento anual es: {media_tasas_crecimiento_INE}%")
print(f"La desviación estándar de las tasas de crecimiento anual es: {desviacion_tasas_crecimiento_INE}%")
```

```python
data = {
    'Categoría': ['Media', 'Desviación Estándar'],
    'InfoJobs': [media_tasas_crecimiento_IJ, desviacion_estandar_IJ],
    'INE': [media_tasas_crecimiento_INE, desviacion_tasas_crecimiento_INE]
}

df = pd.DataFrame(data)
df
```

```python
df.set_index('Categoría', inplace=True)
df
```

```python
ax = df.plot(kind='bar', figsize=(8, 6))
plt.title('Comparación de medias y desviación estándar de las tasas de crecimiento anual')
plt.ylabel('Valores (%)')
plt.xticks(rotation=0)
plt.legend(title='Subgrupos')
plt.tight_layout()

for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', 
                (p.get_x() + p.get_width() / 2, p.get_height()), 
                ha='center', va='center', 
                xytext=(0, 5), 
                textcoords='offset points')

# Mostrar el gráfico
plt.show()
```
