# Comparativa desviaciones típicas salarios INE vs InfoJobs

### Informe InfoJobs 2022

```python
# Informe InfoJobs 2022
salarios_ij_2022
```

```python
salarios_ij_2022_hasta2021 = salarios_ij_2022.iloc[:-1].copy()
salarios_ij_2022_hasta2021
```

```python
salarios_ij_2022_hasta2021.loc[:, 'Tasa de Crecimiento Anual (%)'] = (salarios_ij_2022_hasta2021['Salario Promedio Ofertado (€)'].pct_change() * 100)

tasas_de_crecimiento_IJ = salarios_ij_2022_hasta2021['Tasa de Crecimiento Anual (%)'].dropna()

tasas_de_crecimiento_IJ
```

### Datos INE

```python
salario_INE_nacional_edad = pd.read_excel("databases/salarios/salario_nacional_sexo_y_edad/28189.xlsx")
salario_INE_nacional_edad
```

```python
años = salario_INE_nacional_edad.loc[6, 'Unnamed: 1':'Unnamed: 14'].values
años
```

```python
fila_todas_edades = salario_INE_nacional_edad.loc[7, 'Unnamed: 1':'Unnamed: 14']
fila_todas_edades
```

```python
fila_todas_edades.index = años
fila_todas_edades
```

```python
salarios_2011_2021 = fila_todas_edades.loc[2011:2021]
salarios_2011_2021
```

```python
tasas_crecimiento_anual = salarios_2011_2021.pct_change() * 100
tasas_crecimiento_anual = tasas_crecimiento_anual.dropna()

tasas_crecimiento_anual
```

### Medidas de estadística descriptiva

```python
#InfoJobs

media_tasas_crecimiento_IJ = tasas_de_crecimiento_IJ.mean()
desviacion_estandar_IJ = tasas_de_crecimiento_IJ.std()

print("La media de las tasas de crecimiento anual es:", media_tasas_crecimiento_IJ)
print("Desviación Estándar de las Tasas de Crecimiento Anual:", desviacion_estandar_IJ)
```

```python
#INE

media_tasas_crecimiento_INE = tasas_crecimiento_anual.mean()
desviacion_tasas_crecimiento_INE = tasas_crecimiento_anual.std()

print(f"La media de las tasas de crecimiento anual es: {media_tasas_crecimiento_INE}%")
print(f"La desviación estándar de las tasas de crecimiento anual es: {desviacion_tasas_crecimiento_INE}%")
```

```python
data = {
    'Categoría': ['Media', 'Desviación Estándar'],
    'InfoJobs': [media_tasas_crecimiento_IJ, desviacion_estandar_IJ],
    'INE': [media_tasas_crecimiento_INE, desviacion_tasas_crecimiento_INE]
}

df = pd.DataFrame(data)
df
```

```python
df.set_index('Categoría', inplace=True)
df
```

```python
ax = df.plot(kind='bar', figsize=(8, 6))
plt.title('Comparación de medias y desviación estándar de las tasas de crecimiento anual')
plt.ylabel('Valores (%)')
plt.xticks(rotation=0)
plt.legend(title='Subgrupos')
plt.tight_layout()

for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', 
                (p.get_x() + p.get_width() / 2, p.get_height()), 
                ha='center', va='center', 
                xytext=(0, 5), 
                textcoords='offset points')

# Mostrar el gráfico
plt.show()
```
